### Bagaimana kita dapat  melihat kecenderungan kategori tertentu dalam suatu kolom tertentu akan mengalami attrition atau tidak berdasarkan "tingkat attrition"?
Pertama, kita membuat kolom baru (Add_Attritiom) yang berisi angka biner satu atau dua. Jika pada kolom Attrition bernilai Yes, maka akan diubah menjadi angka satu, jika No diubah menjadi 0. 

```python
def ADD_ATTRITION(x):
    if x["Attrition"]=="Yes":
        baru = 1
    else :
        baru = 0
    return baru
HR["Add_Attrition"] = HR.apply(lambda x : ADD_ATTRITION(x), axis=1 )
HR.head()
```

```python
sns.catplot(x="Department", y="Add_Attrition", hue="Gender", kind="bar", data=HR)

# Dengan bar blot, secara default akan mengeluarkan rata2 nilai, jadi Add_attrition akan merepresentasikan
# "tingkat/rate dari Attrition" berdasarkan kategori tertentu pd kolom tertentu
# Karyawan berjenis kelamin perempuan yang berasal dari departemen HR mempunyai tingkat_attrition lebih besar.
```

```python
# Berasal dari level eduacation manakah karyawan yang mengalami attrition?
sns.catplot(x="Education",  kind="count", hue="Gender", col="Attrition", palette="pastel",data=HR)

# Dengan level education nomor 3, yaitu bachelor, karyawan terbanyak mengalami attrition.
```

```python
# Karyawan dengan level pendidikan nomer satu mempunyai "tingkat_attriction" paling besar
sns.catplot(x="Education", y="Add_Attrition", hue="Gender", kind="bar", height=6, aspect=2,data=HR)
```
